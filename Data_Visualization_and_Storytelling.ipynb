```python
# üõç Superstore Sales Data Analysis with Data Cleaning
#A detailed exploratory data analysis on Superstore data including structured cleaning, transformations, and visual storytelling.
```

```python
# import the library
import pandas as pd
```

```python
# Step 2: Load Data
#Since the dataset contains the characters which cannot be read by using 'utf-8' encoding which is default,we use other encoding method
df = pd.read_csv('Sample - Superstore.csv', encoding='ISO-8859-1')
df.head()
```

```python
# Step 3: Initial Overview
print("Shape of the dataset:", df.shape)
df.info()
df.describe(include='all')
```

```python
# Step 4: Remove Duplicates (if any)
duplicate_count = df.duplicated().sum()
print("Duplicate Rows:", duplicate_count)
df = df.drop_duplicates()
```

```python
# Step 5: Check & Handle Missing Values
# Count of missing values in each column
missing = df.isnull().sum()
print("Missing Values:\n", missing)

# Drop Postal Code (if not relevant or heavily missing)
df = df.drop(columns=['Postal Code'], errors='ignore')
```

```python
# Step 6: Clean Data Types
# Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

# Ensure numeric columns are correct
numeric_cols = ['Sales', 'Quantity', 'Discount', 'Profit']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Drop rows with any NaN after conversions
df.dropna(subset=['Order Date', 'Ship Date', 'Sales', 'Profit'], inplace=True)
```

```python
# Step 7: Validate Categorical Columns
# Check unique values in category-type fields
categorical_cols = ['Region', 'Segment', 'Category', 'Sub-Category', 'Ship Mode']
for col in categorical_cols:
    print(f"\nUnique values in {col}:", df[col].unique())
```

```python
# Step 8: Treat Outliers (Basic Visualization)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
sns.boxplot(data=df[['Sales', 'Profit']])
plt.title("Sales & Profit Outlier Check")
plt.show()

# You can optionally filter out extreme values (e.g., Profit > 10000)
df = df[df['Profit'] < 10000]
```

```python
# Step 9: Feature Engineering
# Add month, year, and profit ratio
df['Month'] = df['Order Date'].dt.month_name()
df['Year'] = df['Order Date'].dt.year
df['Profit Ratio'] = df['Profit'] / df['Sales']
```

```python
## Cleaned Data Overview
#Now that the data is cleaned and structured, we proceed with visualization and business storytelling.
```

```python
## Data Visualization Starts Here
# We will now generate various charts using Seaborn and Matplotlib to uncover insights.
```

```python
# Step 10: Total Sales by Region
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Region', y='Sales', estimator=sum, ci=None)
plt.title('Total Sales by Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 11: Sales and Profit Trend Over Time
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_profit = df.groupby(df['Order Date'].dt.to_period('M'))['Profit'].sum()

plt.figure(figsize=(12, 6))
monthly_sales.plot(label='Sales', marker='o')
monthly_profit.plot(label='Profit', marker='o')
plt.title("Monthly Sales and Profit Trend")
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 12: Sales by Category and Sub-Category
plt.figure(figsize=(14, 6))
sns.barplot(data=df, x='Sub-Category', y='Sales', hue='Category', estimator=sum, ci=None)
plt.title('Sales by Sub-Category and Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 13: Discount vs Profit Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category')
plt.title("Discount vs Profit by Category")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  Step 14: Top 10 Customers by Sales
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_customers.values, y=top_customers.index, palette='viridis')
plt.title("Top 10 Customers by Sales")
plt.xlabel("Sales")
plt.ylabel("Customer")
plt.tight_layout()
plt.show()
```

```python
#  Business Insights Summary

# - **West** region generated the highest sales overall.
# - **Technology** products are highly profitable, while **Office Supplies** show low margins.
# - High **Discounts** often lead to **negative profits**.
# - Year-end months like **November and December** show spikes in sales.
# - A small group of customers accounts for a large portion of revenue.

# These insights can help optimize pricing strategy, stock decisions, and marketing efforts.
```

```python
## Additional Visualization 1: Total Profit by Sub-Category
# This chart helps identify which sub-categories are generating losses or high profits, guiding inventory or pricing decisions.
```

```python
# Group data by sub-category and sum the profit
subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()

# Plot total profit by sub-category
plt.figure(figsize=(12, 6))
sns.barplot(x=subcat_profit.values, y=subcat_profit.index, palette='coolwarm')
plt.title(' Total Profit by Sub-Category')
plt.xlabel('Total Profit ($)')
plt.ylabel('Sub-Category')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
### Insight:
# - Sub-categories with negative bars (e.g., "Tables") are loss-making.
# - Helps identify where business is losing money despite high sales.
```

```python
## üìä Additional Visualization 2: Sales vs Quantity by Sub-Category (Bubble Plot)
# This visualization compares sub-categories based on the number of units sold and the total sales, with bubble size representing profit.
```

```python
# Aggregate sales, quantity, and profit by sub-category
grouped = df.groupby('Sub-Category').agg({
    'Sales': 'sum',
    'Quantity': 'sum',
    'Profit': 'sum'
}).reset_index()

# Plot a bubble chart (scatter) with Quantity vs Sales, bubble size = Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(data=grouped,
                x='Quantity',
                y='Sales',
                size='Profit',
                hue='Sub-Category',
                sizes=(40, 800),
                legend=False)

plt.title(' Sales vs Quantity by Sub-Category (Bubble = Profit)')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
### Insight:
# - Large bubbles = high-profit sub-categories.
# - Bottom-right = high quantity but low sales (e.g., low-value items).
# - Helps find items with volume but not revenue.
```
